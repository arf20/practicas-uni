```python
f(x) = x/(1+3*x^4)
show(f)
```

```python
p(k) = cos((1-k/12)*pi) # 0..12
show(p)
```

```python
X = [ p(k) for k in [0..12]]
X
```

```python
XY = [ (X[i], f(X[i])) for i in [0..12]]
XY
```

```python
r(a, b, x) = f(a) + ((f(b) - f(a))/(b - a))*(x - a)
show(r)
```

```python
L = piecewise([[(X[i], X[i+1]), r(X[i], X[i+1], x)] for i in [0..11]], var=x)
L
```

```python
plot(f, -1, 1) + point2d(XY, color='red') + plot(L(x), (x, -1, 1), color='magenta')
```

```python
plot(abs(diff(f, x, 2)), -1, 1) + plot(4.1, -1, 1, color='red')
```

```python
cota = (4.1/8)*max((X[i+1]-X[i]) for i in [0..11])^2
cota.n()
```

```python
plot(abs(f(x) - L(x)), (x, -1, 1)) + plot(cota, -1, 1, color='red')
```
