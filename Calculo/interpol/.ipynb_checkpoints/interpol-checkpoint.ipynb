```python
g(x) = sin(x) + 2*cos(x)
show(g)
```

```python
Xb = [ 1 + 2*k/5 for k in [0..5]]
XYb = [ (Xb[i], g(Xb[i])) for i in [0..5] ]
```

```python
Q(x) = RR['x'].lagrange_polynomial(XYb)
show(Q)
```

```python
plot(g, -2, 5) + plot(Q, -2, 5, color='magenta') + point2d(XYb)
```

```python
plot(abs(diff(g, x, 6)), 1, 3) + plot(1.95, 1, 3, color='red')
```

```python
cota = 1.95/factorial(6)
cota
```

```python
plot(abs(g(x) - Q(x)), x, 1, 3) + plot(cota, 1, 3, color='red')
```
