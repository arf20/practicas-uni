```python
f(x) = e^(-x^2)
n = 10
a = -2
b = 1
show(f)
```

```python
X = [ (a + k*(b-a)/(2*n)) for k in [0..2*n] ]
X
```

```python
simpson = ((b-a)/(6*n))*(f(a) + 4*sum(f(X[2*i-1]) for i in [1..n]) + 2*sum(f(X[2*i]) for i in [1..n-1]) + f(b))
simpson.n()
```

```python
cota = 12.1 * (b-a)^5/(180*(2*n)^4)
cota
```

```python
abs(integral(f, x, a, b) - simpson).n() < cota
```
