### d. calcular raiz error<10^8

```python
a0 = a
b0 = b
c0 = (a0+b0)/2
N = 23
i = 0
while i <= N:
	if (f(a0)*f(c0) < 0):
		b0 = c0
		c0 = (a0+b0)/2
	else:
		a0 = c0
		c0 = (a0+b0)/2
	i+=1
c0
```
